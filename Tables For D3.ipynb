```python
import pandas as pd
```

```python
co2_emissions = pd.read_csv('agrofood_co2_emission_cleaned.csv')
co2_emissions.head()
```

```python
emissions_by_year = co2_emissions.groupby('Year')['total_emission'].sum().reset_index()
emissions_by_year.head()
```

```python
avg_temp_by_year = co2_emissions.groupby('Year')['Average Temperature °C'].mean().reset_index()
avg_temp_by_year.head()
```

```python
avg_temp_by_year.to_csv('avg_temp.csv', index=False)
emissions_by_year.to_csv('emissions_by_year.csv', index=False)
```
